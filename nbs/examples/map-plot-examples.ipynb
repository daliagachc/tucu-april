```python
import cartopy
import matplotlib
import matplotlib.pyplot as plt 
```

```python
lom = -70
loM = -67
lam = -18
laM = -15
```

```python
fsize = (9,10)
```

```python
ccrs = cartopy.crs
```

```python
crs = ccrs.Mercator()
fig = plt.figure(figsize=fsize)
geo_axes = plt.axes(projection=crs)

states = cartopy.feature.NaturalEarthFeature(category='cultural', 
                             scale='110m', # 10 or 50 works as well
                             facecolor='none',
                             name='admin_0_countries')

geo_axes.add_feature(states, linewidth=2.0, edgecolor='black', zorder=2)

geo_axes.stock_img()


```

```python
crs = ccrs.Mercator()
fig = plt.figure(figsize=fsize)
geo_axes = plt.axes(projection=crs)

states = cartopy.feature.NaturalEarthFeature(category='cultural', 
                             scale='50m', 
                             facecolor='none',
                             name='admin_0_countries')

geo_axes.add_feature(states, linewidth=2.0, edgecolor='black',facecolor='none')

geo_axes.stock_img()

geo_axes.set_extent((lom,loM,lam,laM))
```

```python
crs = ccrs.Mercator()
fig = plt.figure(figsize=fsize)
geo_axes = plt.axes(projection=crs)
geo_axes.set_extent((lom,loM,lam,laM))
geo_axes.background_img(name='BM', resolution='high')

states = cartopy.feature.NaturalEarthFeature(category='cultural', 
                             scale='50m', 
                             facecolor='none',
                             name='admin_0_countries')

geo_axes.add_feature(states, linewidth=2.0, edgecolor='black',facecolor='none')
geo_axes.add_feature(cartopy.feature.LAND)
# geo_axes.add_image(request, 8, interpolation='spline36')
# geo_axes.stock_img()

```

```python
import matplotlib.pyplot as plt

import cartopy.crs as ccrs
from cartopy.io import shapereader
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER

import cartopy.io.img_tiles as cimgt

extent = [lom,loM,lam,laM]

request = cimgt.OSM()

fig = plt.figure(figsize=(9, 13))
ax = plt.axes(projection=request.crs)
gl = ax.gridlines(draw_labels=True, alpha=0.2)
gl.xlabels_top = gl.ylabels_right = False
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER

ax.set_extent(extent)

ax.add_image(request, 8, interpolation='spline36')
plt.show()
```

```python
import matplotlib.pyplot as plt

import cartopy.crs as ccrs
from cartopy.io import shapereader
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER

import cartopy.io.img_tiles as cimgt

extent = [lom,loM,lam,laM]

request = cimgt.GoogleTiles(style='satellite',
                           
                          url= ('https://mts0.google.com/vt/lyrs=y'
                      '@177000000&hl=en&src=api&x={x}&y={y}&z={z}&s=G'))

fig = plt.figure(figsize=(9, 13))
ax = plt.axes(projection=request.crs)
gl = ax.gridlines(draw_labels=True, alpha=0.2)
gl.xlabels_top = gl.ylabels_right = False
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER

ax.set_extent(extent)

ax.add_image(request, 8, interpolation='spline36')
plt.show()
```
